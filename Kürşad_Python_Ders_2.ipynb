```python
# Galibiyetin 2 puan olduğu yılları normalleştirmek için tekrar puan hesaplaması yapalım.
```

```python
lig['puan_duzeltilmis']=(lig['galibiyet']*3)+(lig['beraberlik']*1)
```

```python
# En Yüksek Puanla Şampiyon Olan Takım
maxpuan=lig['puan_duzeltilmis'].max()
lig[lig['puan_duzeltilmis']==maxpuan]
```

```python
# En Düşük Puanla Şampiyon Olan Takım
minpuan=lig['puan_duzeltilmis'].min()
lig[lig['puan_duzeltilmis']==minpuan]
```

```python
# En Yüksek Puan Ortalaması ile Şampiyon Olan Takım
lig['ort_puan']=lig['puan_duzeltilmis']/lig['oyun']

maxortpuan=lig['ort_puan'].max()
lig[lig['ort_puan']==maxortpuan]
```

```python
# En Az Puan Ortalaması ile Şampiyon Olan Takım
minortpuan=lig['ort_puan'].min()
lig[lig['ort_puan']==minortpuan]
```

```python
# Şampiyon Olan Takımların Maç Başı Ortalama Gol Sayıları

lig['ort_gol']=lig['atilan_gol']/lig['oyun']
```

```python
#Süperlig Tarihinin Şampiyonlarının Maç Başına Ortalama Puanları
lig['ort_puan'].mean()
```
